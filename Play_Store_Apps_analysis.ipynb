# **What is Exploratory Data Analysis?**
Exploratory data analysis (EDA) is used by data scientists to examine data sets for patterns and anomalies (outliers), to create hypotheses based on our comprehension of the dataset, and to summarize their key properties. Data visualization techniques are frequently utilized in this process. In any project involving data analysis or data science, it is a crucial stage. It helps in deciding how to best manipulate data sources to obtain the desired results.

In order to better comprehend the data and make it more visually attractive and appealing, EDA involves producing summary statistics for the dataset's numerical data.

The following are the various steps involved in the EDA process:
1. <b>Problem Statement</b> - We'll come up with ideas and analyze the provided data. We will investigate the characteristics it has and attempt to conduct a philosophical examination of their significance and relevance to this issue.
2. <b>Hypothesis</b> - Following our examination of the data base's features, we will formulate some fundamental hypotheses on which to base our exploration of the data's potential outcomes.
3. <b>Univariate Analysis</b> - It is the most straightforward method of data analysis. In order to start, we would choose just one attribute and thoroughly examine it. It deals with no co-relation whatsoever, and its primary goal is to describe. Data is taken, summarized, and patterns are discovered within the data.
4. <b>Bivariate Analysis</b> - Cause and the connection between the two traits are the subjects of this analysis. We'll make an effort to comprehend how traits are related to one another.
5. <b>Multivariate Analysis</b> - this is what's done.when it's necessary to investigate multiple variables at once.
5. <b>Data Cleaning</b> - The dataset will be cleaned, and missing data, outliers, and categorical variables will be handled.
6. <b>Testing Hypothesis</b> - We will determine whether our data complies with the assumptions needed by the majority of multivariate approaches.
