```python
df1_noduplicates = df1.drop_duplicates()
```

```python
New_df1.Rating
```

```python
#New_df1.Installs.sort_values(ascending=False).head(20)
final_df = New_df1.copy()
final_df = final_df.drop(columns=['index'])
special_characters=[',','+']
shapes = final_df.shape
lengths = shapes[0]
type(length)
for i in range(0,lengths+1):
    for a in special_characters:
        final_df.loc[i,'Installs'] = final_df.loc[i,'Installs'].replace(a,"")
    final_df.loc[i,'Installs'] = int(final_df.loc[i,'Installs']) 
    print(final_df.loc[i,'Installs'])
    print(type(final_df.loc[i,'Installs']))
    
```

```python
final_df
```

```python
length = final_df.shape
type(length[0])
```

```python
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
Pop_Cat = final_df.groupby('Category')['Installs'].mean()
```

```python
Pop_Cat = Pop_Cat.to_frame()
```

```python
Pop_Cat = Pop_Cat.sort_values(by='Installs',ascending=False)
```

```python
Pop_Cat.info()
```

```python
Pop_Cat
```

```python
plt.figure(figsize=(15,15))
sns.set_theme(style="white", context="talk")
sns.barplot(data=Pop_Cat,x='Installs', y=Pop_Cat.index)
plt.title('Most Installed Categories'.title())
plt.show()
```

## Finding the top most installed apps (top 10 and top 20) Plot them.

```python
Top_ten = New_df1.groupby(New_df1['App'])['App'].count().sort_values().tail(10)
Top_ten.rename(index={'Duolingo: Learn Languages Free': 'Duolingo'},inplace=True)

Top_ten
```

```python
Top_ten.shape
```

```python
Top_ten.index
```

```python
plt.figure(figsize=(11,5))
plt.bar(Top_ten.index, 
        Top_ten.values,
        color = 'purple')
plt.xlabel('Name of the app')
plt.ylabel('Number of downloads')
plt.xticks(rotation=80)
plt.title('Top 10 apps'.title())
plt.show()
```

```python
Top_twenty = New_df1.groupby(New_df1['App'])['App'].count().sort_values().tail(20)
Top_twenty.rename(index={'Sniper 3D Gun Shooter: Free Shooting Games - FPS': 'Sniper 3D Gun Shooter'},inplace=True)
Top_twenty.rename(index={'Duolingo: Learn Languages Free': 'Duolingo'},inplace=True)
Top_twenty
```

```python
plt.figure(figsize=(12,6))
plt.bar(Top_twenty.index, 
        Top_twenty.values,
        color = 'purple')
plt.xlabel('Name of the app')
plt.ylabel('Number of Downloads')
plt.xticks(rotation=80)
plt.title('Top 20 apps'.title())
plt.show()
```

```python
#Most installed categories
Most_categories =  New_df1.groupby(['Category']).aggregate({'App':'count'})
Most_categories = Most_categories.sort_values(by=['App'], ascending=False)
Most_categories
```

```python
Most_categories['App'].sum()
```

```python
#plt.figure(figsize=(20,20))
plot = Most_categories.head(15).plot.pie(y='App', autopct='%1.0f%%', figsize=(10, 10))
```

```python
df_z_scaled = Most_categories.copy()
df_z_scaled.dtypes
```

```python
#df_z_scaled['App'] = df_z_scaled['App'].astype('float').div(8886).round(2)
df_z_scaled['App'] = df_z_scaled.loc[:].div(df_z_scaled.sum()).div(1/100).round(6)
```

```python
df_z_scaled['App'].dtypes
```

```python
df_z_scaled
```

```python
plt.figure(figsize=(15,15))
sns.set_theme(style="white", context="talk")
sns.barplot(data=Most_categories,x='App', y=Most_categories.index)
plt.title('Most Installed Categories'.title())
plt.show()
```
